# PDF loader

```python
from langchain.document_loaders import PyPDFLoader
loader = PyPDFLoader("National AI Policy Consultation Draft V1.pdf")
pages = loader.load()
```

```python
#Check the length of pages
len(pages)
```

```python
page = pages[0]
```

```python
page
```

```python
# Print the first 500 words (with space)
print(page.page_content[0:500])
```

```python
#Check the source of data
page.metadata["source"]
```

```python
#Print the metadata page number
#hint use metadata and page
page.metadata
```
